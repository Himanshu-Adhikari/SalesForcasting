```python
#Importing libraries
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/supplement.csv")
dataset
```

```python
dataset.isnull().sum()
```

```python
import plotly.express as px
```

```python
pie = dataset["Store_Type"].value_counts()
store = pie.index
orders = pie.values
fig = px.pie(dataset , values=orders , names = store)
fig.show()
```

```python
pie = dataset["Location_Type"].value_counts()
location = pie.index
orders = pie.values
fig = px.pie(dataset , values=orders , names = location)
fig.show()
```

```python
pie = dataset["Region_Code"].value_counts()
region = pie.index
orders = pie.values
fig = px.pie(dataset , values=orders , names = region)
fig.show()
```

```python

pie = dataset["Discount"].value_counts()
discount = pie.index
orders = pie.values
fig = px.pie(dataset , values=orders , names = discount)
fig.show()
```

```python
dataset["Discount"] = dataset["Discount"].map({"Yes":1 , "No":0})
dataset
```

```python
dataset["Store_Type"] = dataset["Store_Type"].map({"S1":1 ,"S2":2,"S3":3,"S4":4})
dataset
```

```python
dataset["Location_Type"] = dataset["Location_Type"].map({"L1":1 ,"L2":2,"L3":3,"L4":4,"L5":5})
dataset
```

```python
x = np.array(dataset[["Store_Type","Location_Type","Holiday","Discount"]])
y = np.array(dataset["#Order"])
```

```python
x
```

```python
y
```

```python
#Building the ML Model
from sklearn.model_selection import train_test_split as tts
```

```python
x_train, x_test, y_train, y_test = tts(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
#importing the model that we are going to implement
import lightgbm as ltb
model = ltb.LGBMRegressor()
```

```python
#passing our test set inside the model
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```
